```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
raw_data = pd.read_csv('Country clusters standardized.csv', index_col = ['Country'] ) # Here we specify that the data 
# should be indexed using the Country
```

```python
data = raw_data.copy()
data = data.drop(['Language'], axis = 1)
data
```

```python
# Now we plot the dendrogram
sns.clustermap(data, cmap = 'mako')
```

```python
# This heatmap shows us that Australia is far from other countries based on the latitude, hence the black color. 
# Also, Canada aand USA are close based on the longitude same as the european countries. However, Australia is distant
# from these countries based on the longitude
```

```python
# From the diagram above, we can tell that USA and Canada belong to the same cluster. Same as Germeny and France and
# Ausralia is in a cluster of its own
# Should we want a cluster of two, we can see that Australia is the outlier and other Countries will form their own
# cluster.
```

#### Let's try it using another dataset

```python
new_data = pd.read_csv('3.12. Example.csv')
new_data
```

```python
new_data_scaled = new_data.copy()
```

```python
# We standardize the data
from sklearn import preprocessing
new_data_scaled = preprocessing.scale(new_data_scaled)
scaled_data = pd.DataFrame(data = new_data_scaled, columns = [['Satisfaction','Loyalty']])
scaled_data 
```

```python
processed_scaled = scaled_data.rename_axis('Customers')
processed_scaled
```

```python
sns.clustermap(processed_scaled, cmap = 'mako')
```

```python
# From this heat map, we can spot 4 clusters(bottom-up). The segment of customers that are loyal but not satisfied(fans), not loyal
# and not satisfied(alienated), not loyal but satisfied(roamers) and finally, loyal and satisfied(Fans)
```
