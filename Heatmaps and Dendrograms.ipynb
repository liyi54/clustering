```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
raw_data = pd.read_csv('Country clusters standardized.csv', index_col = ['Country'] ) # Here we specify that the data 
# should be indexed using the Country
```

```python
data = raw_data.copy()
data = data.drop(['Language'], axis = 1)
data
```

```python
# Now we plot the dendrogram
sns.clustermap(data, cmap = 'mako')
```

```python
# This heatmap shows us that Australia is far from other countries based on the latitude, hence the black color. 
# Also, Canada aand USA are close based on the longitude same as the european countries. However, Australia is distant
# from these countries based on the longitude
```

```python
# From the diagram above, we can tell that USA and Canada belong to the same cluster. Same as Germeny and France and
# Ausralia is in a cluster of its own
# Should we want a cluster of two, we can see that Australia is the outlier and other Countries will form their own
# cluster.
```
